```python
#Que 3) Explain the following terms:
# The difference in np.array, np.asarray and np.asanyarray
# The difference between Deep copy and shallow copy.

# These functions are all used to create NumPy arrays from various input data types.

# A) np.array(object, dtype=None, copy=True, order='K', subok=False, ndmin=0)
# Converts the input object into a NumPy array.
# Creates a copy of the input by default (copy=True).
# Offers more flexibility with parameters like dtype, order, subok, and ndmin for fine-grained control over the output array.

# B) np.asarray(a, dtype=None, order=None):
#Converts the input to an ndarray.
#Does not create a copy if the input is already an ndarray or a subclass of ndarray.
#Less flexible than np.array as it has fewer parameters.

# C) np.asanyarray(a, dtype=None, order=None):
#Similar to np.asarray but is more permissive in terms of input types.
#It tries to convert the input to an ndarray without copying if possible, but it can also create a copy if necessary.

# Key Differences:
# Function	           Copy	             Flexibility
# np.array	       Yes (default)	        High
# np.asarray	   No (if possible)	       Medium
# np.asanyarray	   No (if possible)	        High

# When to use which:
# Use np.array when you need full control over the output array and want to ensure a copy is created.
# Use np.asarray when you want to avoid unnecessary copying and the input is likely already an ndarray.
# Use np.asanyarray when you are unsure about the input type and want a flexible function that tries to avoid copying.


# Deep Copy and Shallow Copy
#Shallow Copy:
#Creates a new object but copies only the references to the original object's data.
#Changes to the original object will be reflected in the copy.
#In NumPy, creating a view of an array using slicing or np.view() is a shallow copy.

#Deep Copy:
#Creates a new object and recursively copies all data from the original object.
#Changes to the original object will not affect the copy.
#In NumPy, using np.copy() or copy.deepcopy() creates a deep copy.

#Example:
import numpy as np
import copy

# Shallow copy
a = np.array([1, 2, 3])
b = a.view()
b[0] = 4
print(a)
```

```python
# Deep copy
c = np.copy(a)
c[0] = 5
print(a)
```

```python
# Que 5)Create a NumPy array with random integers between 1 and 10 of shape(5,6). After creating the array perform the following operations.
#  a)Extract all even integers from array.
#  b)Extract all odd integers from array.
import numpy as np

# Create a random array of shape (5, 6) with integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

print("Original array:\n", random_array)

# Extract even integers
even_integers = random_array[random_array % 2 == 0]
print("\nEven integers:\n", even_integers)

# Extract odd integers
odd_integers = random_array[random_array % 2 != 0]
print("\nOdd integers:\n", odd_integers)
```

```python
# Que 6)Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations,
#  a) Find the indices of the maximum values along each depth level(third axis).
#  b) Perform element-wise multiplication of between both array.
import numpy as np

# Create a 3D NumPy array of shape (3, 3, 3) with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(3, 3, 3))
print("Original array:\n", random_array)

# Find the indices of maximum values along the third axis
max_indices = np.argmax(random_array, axis=2)
print("\nIndices of maximum values along each depth level:\n", max_indices)

# Perform element-wise multiplication of the original array and the array of maximum indices
result = random_array * max_indices
print("\nElement-wise multiplication result:\n", result)
```

```python
# Que 15) b) Overlay a line plot representing the normal distribution's probability density function (PDF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create 1000 samples from a normal distribution
data = np.random.randn(1000)

# Calculate histogram
hist, bins = np.histogram(data, bins=30, density=True)

# Calculate PDF of the normal distribution
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x)

# Plot histogram and PDF
plt.hist(data, bins=bins, density=True, alpha=0.6, label='Histogram')
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram and PDF of Normal Distribution')
plt.legend()

plt.show()
```

```python
# Que 17) Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
#         origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Create two random arrays
x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)

# Create a DataFrame with quadrant information
df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Quadrant': np.where((x >= 0) & (y >= 0), 'Quadrant 1',
                np.where((x < 0) & (y >= 0), 'Quadrant 2',
                np.where((x < 0) & (y < 0), 'Quadrant 3', 'Quadrant 4')))
})

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='viridis')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrants')

# Show the plot
plt.show()
```

```python
# Que 19) Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

# !pip install bokeh
```

```python
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
import random

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(1, 100) for _ in categories]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add a bar renderer
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Show the plot
show(p)
```
